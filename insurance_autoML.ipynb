```python
import os
import pandas as pd
from pycaret.regression import *
```

```python
FEATURES = ['age', 'bmi', 'children', 'smoker']
TARGET = 'charges'
```

```python
data = pd.read_csv('../data/insurance.csv')
data = data[[*FEATURES,TARGET]]
data
```

```python
s = setup(data, target='charges', silent=True, session_id=2023)
```

```python
best_model = compare_models()
```

```python
best_model
```

```python
tuned_model = tune_model(best_model, fold=5, n_iter=100)
```

```python
evaluate_model(tuned_model)
```

```python
final_model = finalize_model(tuned_model)
```

```python
os.makedirs('../app', exist_ok=True)
create_api(final_model, '../app/api_insurance')
```
